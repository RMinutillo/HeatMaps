```python

# Coloca tu clave API de Google Maps aquí
api_key = 'AIzaSyDKHDfzW8QG-KBoEjn16bPRDzSkuMd4hjg'

# Crea un cliente de Google Maps
gmaps = googlemaps.Client(key=api_key)

# Especifica las coordenadas de inicio y fin
start_coords = (-34.714938016373594, -58.33892002268164)  # Nueva York
end_coords = (-34.702172421921446, -58.38656017148223)    # Los Ángeles

# Obtiene la dirección y la duración estimada del viaje
directions_result = gmaps.directions(start_coords, end_coords, mode="driving", departure_time="now")

# Extrae las coordenadas de la ruta
polyline_points = directions_result[0]['overview_polyline']['points']

# Decodifica la polyline a una lista de coordenadas
route_coords = googlemaps.convert.decode_polyline(polyline_points)

# Asegúrate de que las coordenadas estén en el formato correcto
route_coords = [(coord['lat'], coord['lng']) for coord in route_coords]

# Crea un mapa con Folium
map_route = folium.Map(location=start_coords, zoom_start=5)

# Agrega una línea a la ruta en el mapa
folium.PolyLine(locations=route_coords, color='blue').add_to(map_route)

# Agrega marcadores para el inicio y el final
folium.Marker(location=start_coords, popup='Inicio', icon=folium.Icon(color='green')).add_to(map_route)
folium.Marker(location=end_coords, popup='Fin', icon=folium.Icon(color='red')).add_to(map_route)

map_route
```
