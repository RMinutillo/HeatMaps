```python
import folium
from folium.plugins import MiniMap
```

```python
popuptext = '<b>Casa Domi y Rami</b>'

casa = folium.Map(location=[-34.75711565330841, -58.29451049531041],zoom_start=16)

folium.Marker(location=[-34.75711565330841, -58.29451049531041],popup=popuptext).add_to(casa)

folium.Circle(location=[-34.75711565330841, -58.29451049531041],color='purple',fill_color='red', radius=40,weigth=4,fill_opacity=0.5,tooltip='Casita').add_to(casa)

minimap = MiniMap()

casa.add_child(minimap)

casa
```
