```python
# Libreries
import pandas as pd
import pyvis
from pyvis.network import Network
```

```python
# Import csv to df
df_stormofswords = pd.read_csv('stormofswords.csv')
df_datasprint2_nodos = pd.read_csv('data_sprint2.csv')
```

```python
#Inspection df
#print(df_stormofswords.columns)
#print(df_stormofswords.describe)
#print(df_stormofswords.isnull().sum())

print(df_datasprint2_nodos.columns)
print(df_datasprint2_nodos.describe)
print(df_datasprint2_nodos.isnull().sum())

```

```python
#df_datasprint2_nodos.dropna(inplace=True)
df_datasprint2_nodos.isnull().sum()
df_datasprint2_nodos.duplicated().sum()

# Renombrar las columnas eliminando espacios en blanco
df_datasprint2_nodos = df_datasprint2_nodos.rename(columns=lambda x: x.strip())
```

```python
# Crear un objeto de red PyVis
grafo = Network(notebook=True, cdn_resources="remote" ,height='800px', width='100%')

# Agrupar los datos por "Brigade Member"
grouped_data = df_datasprint2_nodos.groupby('Brigade Member')

# Agregar nodos
for brigade_member, group in grouped_data:
    for index, row in group.iterrows():
        grafo.add_node(row['Name registered'], title=f"Nombre: {row['Name registered']}<br>Dirección: {row['Address']}<br>Teléfono: {row['Phone']}<br>Fecha de Registro: {row['Date registered']}")

# Agregar conexiones
for brigade_member, group in grouped_data:
    group_list = group['Name registered'].tolist()
    for i in range(len(group_list)):
        for j in range(i + 1, len(group_list)):
            grafo.add_edge(group_list[i], group_list[j])

# Visualizar el grafo
grafo.show("pepe.html")


```
